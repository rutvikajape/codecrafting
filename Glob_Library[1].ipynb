```python
import tkinter as tk
from tkinter import ttk
import glob
import os
from typing import List

class Glob:
  

  def _init_(self, pattern: str):
   
    self.pattern = pattern

  def expand(self) -> List[str]:
    
    return glob.glob(self.pattern, recursive=True)

  def expand_with_basename(self) -> List[str]:
    
    return [os.path.basename(path) for path in self.expand()]

  def expand_with_directory(self) -> List[str]:
   
    return [os.path.dirname(path) for path in self.expand()]

  def count(self) -> int:
   
    return len(self.expand())

  def exists(self) -> bool:
    
    return self.count() > 0

  def is_directory(self) -> bool:
   
    return os.path.isdir(self.pattern)

  def is_file(self) -> bool:
   
    return os.path.isfile(self.pattern)

  def is_absolute(self) -> bool:
   
    return os.path.isabs(self.pattern)

  def is_relative(self) -> bool:
    
    return not self.is_absolute()

  def get_size(self) -> int:
   
    total_size = 0
    for path in self.expand():
      if os.path.isfile(path):
        total_size += os.path.getsize(path)
    return total_size

  def get_modified_time(self) -> int:
    
    latest_time = 0
    for path in self.expand():
      if os.path.isfile(path):
        modified_time = os.path.getmtime(path)
        if modified_time > latest_time:
          latest_time = modified_time
    return latest_time

  def get_created_time(self) -> int:
    
    latest_time = 0
    for path in self.expand():
      if os.path.isfile(path):
        created_time = os.path.getctime(path)
        if created_time > latest_time:
          latest_time = created_time
    return latest_time

  def get_access_time(self) -> int:
   
    latest_time = 0
    for path in self.expand():
      if os.path.isfile(path):
        access_time = os.path.getatime(path)
        if access_time > latest_time:
          latest_time = access_time
    return latest_time

def expand_and_display(pattern):
 
  g = Glob(pattern)
  matching_files = g.expand()
  listbox.delete(0, tk.END)
  for file in matching_files:
    listbox.insert(tk.END, file)

# Create main window
window = tk.Tk()
window.title("Glob Pattern Expander")

# Create label and entry for glob pattern
pattern_label = tk.Label(window, text="Enter a glob pattern:")
pattern_label.pack()
pattern_entry = tk.Entry(window)
pattern_entry.pack()

# Create button to expand the pattern
expand_button = tk.Button(window, text="Expand", command=lambda: expand_and_display(pattern_entry.get()))
expand_button.pack()

# Create listbox to display results
listbox = tk.Listbox(window)
listbox.pack()

window.mainloop()
```
